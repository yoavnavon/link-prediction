```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

# Dynamobi

### Random 01-02/08

```python
heuristic = pd.read_csv('results/dynamobi/17_random_heuristic.csv')
node2vec = pd.read_csv('results/dynamobi/17_random_node2vec.csv')
deepwalk = pd.read_csv('results/dynamobi/17_random_deepwalk.csv')
```

```python
heuristic.head(5)
```

```python
fig, ax = plt.subplots(2, 2, sharex='col', figsize=(20, 10))
# Precision
ax[0][0].plot('size','precision','-o', data=heuristic[heuristic['model'] == 'rf'])
ax[0][0].plot('size','precision','-o', data=node2vec[node2vec['model'] == 'rf'])
ax[0][0].plot('size','precision','-o', data=deepwalk[deepwalk['model'] == 'rf'])
ax[0][0].set_xlabel('N edges')
ax[0][0].set_ylabel('Precision')
ax[0][0].legend(['traditional','node2vec','deepwalk'])

# Recall
ax[0][1].plot('size','recall','-o', data=heuristic[heuristic['model'] == 'rf'])
ax[0][1].plot('size','recall','-o', data=node2vec[node2vec['model'] == 'rf'])
ax[0][1].plot('size','recall','-o', data=deepwalk[deepwalk['model'] == 'rf'])
ax[0][1].set_xlabel('N edges')
ax[0][1].set_ylabel('Recall')

# ROC AUC
ax[1][0].plot('size','roc_auc','-o', data=heuristic[heuristic['model'] == 'rf'])
ax[1][0].plot('size','roc_auc','-o', data=node2vec[node2vec['model'] == 'rf'])
ax[1][0].plot('size','roc_auc','-o', data=deepwalk[deepwalk['model'] == 'rf'])
ax[1][0].set_xlabel('N edges')
ax[1][0].set_ylabel('ROC AUC')

# F1 Score
ax[1][1].plot('size','f1_score','-o', data=heuristic[heuristic['model'] == 'rf'])
ax[1][1].plot('size','f1_score','-o', data=node2vec[node2vec['model'] == 'rf'])
ax[1][1].plot('size','f1_score','-o', data=deepwalk[deepwalk['model'] == 'rf'])
ax[1][1].set_xlabel('N edges')
ax[1][1].set_ylabel('F1 Score')
plt.show()

fig, ax = plt.subplots(1, 1, sharex='col', figsize=(10, 5))
ax.plot('size','accuracy','-o', data=heuristic[heuristic['model'] == 'rf'])
ax.plot('size','accuracy','-o', data=node2vec[node2vec['model'] == 'rf'])
ax.plot('size','accuracy','-o', data=deepwalk[deepwalk['model'] == 'rf'])
ax.set_xlabel('N edges')
ax.set_ylabel('Accuracy')
ax.legend(['traditional','node2vec','deepwalk'])
plt.show()
```
