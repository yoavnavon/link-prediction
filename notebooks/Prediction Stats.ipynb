```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dynamobi = pd.read_csv('results/dynamobi/28_prediction-stats.csv')
youtube = pd.read_csv('results/youtube/26_prediction-stats.csv')
hepph = pd.read_csv('results/hepph/26_prediction-stats.csv')
twitter = pd.read_csv('results/twitter/26_prediction-stats.csv')
```

```python
dynamobi
```

```python
fig, ax = plt.subplots(2,2,figsize=(15,10))
split = 'POS'
ax1 = sns.lineplot(x='Size',y='Hub',hue='Method',data=dynamobi[dynamobi.Split==split],style="Method",markers=True,dashes=False,ax=ax[0][0])
ax1.set_title('Dynamobi')
ax2 = sns.lineplot(x='Size',y='Hub',hue='Method',data=youtube[youtube.Split==split],style="Method",markers=True,dashes=False,ax=ax[0][1])
ax2.set_title('Youtube')
ax3 = sns.lineplot(x='Size',y='Hub',hue='Method',data=hepph[hepph.Split==split],style="Method",markers=True,dashes=False,ax=ax[1][0])
ax3.set_title('HepPh')
ax4 = sns.lineplot(x='Size',y='Hub',hue='Method',data=twitter[twitter.Split==split],style="Method",markers=True,dashes=False,ax=ax[1][1])
ax4.set_title('Twitter')
plt.show()
```
