```python
import pandas as pd
import json
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pd.set_option('display.max_columns', None)
```

```python
metrics_dynamobi = json.load(open('results/dynamobi/metrics.json','r'))
df_dynamobi = pd.DataFrame(metrics_dynamobi).transpose()

metrics_youtube = json.load(open('results/youtube/metrics.json','r'))
df_youtube = pd.DataFrame(metrics_youtube).transpose()

metrics_hepph = json.load(open('results/hepph/metrics.json','r'))
df_hepph = pd.DataFrame(metrics_hepph).transpose()
```

```python
df_dynamobi
```

```python
fig, ax = plt.subplots(5, 2, figsize=(20, 20))
index_dynamobi = df_dynamobi.index
index_youtube = df_youtube.index
index_hepph = df_hepph.index
legend = ['dynamobi','youtube','hepph']

ax[0][0].plot(index_dynamobi,df_dynamobi['avg-out-degree'],'-o')
ax[0][0].plot(index_youtube,df_youtube['avg-out-degree'],'-o')
#ax[0][0].plot(index_hepph,df_hepph['avg-out-degree'],'-o')
ax[0][0].legend(legend)
ax[0][0].set_title('Avg Out Degree')

ax[0][1].plot(index_dynamobi,df_dynamobi['max-out-degree'],'-o')
ax[0][1].plot(index_youtube,df_youtube['max-out-degree'],'-o')
#ax[0][1].plot(index_hepph,df_hepph['max-out-degree'],'-o')
ax[0][1].legend(legend)
ax[0][1].set_title('Max Out Degree')

ax[1][0].plot(index_dynamobi,df_dynamobi['avg-neighboor-degree'],'-o')
ax[1][0].plot(index_youtube,df_youtube['avg-neighboor-degree'],'-o')
#ax[1][0].plot(index_hepph,df_hepph['avg-neighboor-degree'],'-o')
ax[1][0].legend(legend)
ax[1][0].set_title('Avg Neighboor Degree')

ax[1][1].plot(index_dynamobi,df_dynamobi['max-neighboor-degree'],'-o')
ax[1][1].plot(index_youtube,df_youtube['max-neighboor-degree'],'-o')
#ax[1][1].plot(index_hepph,df_hepph['max-neighboor-degree'],'-o')
ax[1][1].legend(legend)
ax[1][1].set_title('Max Neighboor Degree')

ax[2][0].plot(index_dynamobi,df_dynamobi['count-scc'],'-o')
ax[2][0].plot(index_youtube,df_youtube['count-scc'],'-o')
#ax[2][0].plot(index_hepph,df_hepph['count-scc'],'-o')
ax[2][0].legend(legend)
ax[2][0].set_title('Count SCC')

ax[2][1].plot(index_dynamobi,df_dynamobi['avg-scc'],'-o')
ax[2][1].plot(index_youtube,df_youtube['avg-scc'],'-o')
#ax[2][1].plot(index_hepph,df_hepph['avg-scc'],'-o')
ax[2][1].legend(legend)
ax[2][1].set_title('Avg SCC size')

ax[3][0].plot(index_dynamobi,df_dynamobi['count-wcc'],'-o')
ax[3][0].plot(index_youtube,df_youtube['count-wcc'],'-o')
#ax[3][0].plot(index_hepph,df_hepph['count-wcc'],'-o')
ax[3][0].legend(legend)
ax[3][0].set_title('Count WCC')

ax[3][1].plot(index_dynamobi,df_dynamobi['avg-wcc'],'-o')
ax[3][1].plot(index_youtube,df_youtube['avg-wcc'],'-o')
#ax[3][1].plot(index_hepph,df_hepph['avg-wcc'],'-o')
ax[3][1].legend(legend)
ax[3][1].set_title('Avg WCC size')

ax[4][0].plot(index_dynamobi,df_dynamobi['avg-clustering'],'-o')
ax[4][0].plot(index_youtube,df_youtube['avg-clustering'],'-o')
#ax[4][0].plot(index_hepph,df_hepph['avg-clustering'],'-o')
ax[4][0].legend(legend)
ax[4][0].set_title('Avg Clustering')


plt.show()
```
